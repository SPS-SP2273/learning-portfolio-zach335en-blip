```python
from matplotlib import pyplot as plt
import numpy as np
```

# Exercise 1 (Fill in the blanks)

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc= 'lower right')
plt.show()
```

# Exercise 2 (A simple mathematical plot)  

```python
x = np.linspace(0,50, 1000)
y = np.e ** (-1 * x/10) * np.sin(x)
plt.plot(x, y, label = 'y-values')
plt.xlabel('x', c= 'darkorange')
plt.ylabel('y', c = 'darkorange', rotation = 0)
plt.title(r'$y=e^{-x/10} sin{(x)}$')
plt.grid(alpha=.25)
plt.legend(loc= 'lower right')
plt.rc('axes', titlesize = 15)
plt.rc('figure', titlesize = 20)
plt.ylim(-1, 1)
plt.show()
plt.tight_layout()
```
